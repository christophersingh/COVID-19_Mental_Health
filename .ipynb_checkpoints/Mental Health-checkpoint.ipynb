```python
import pandas as pd
from sodapy import Socrata

client = Socrata("data.cdc.gov", None)

results = client.get("8pt5-q6wp", limit=3000)

# Convert to pandas DataFrame
df = pd.DataFrame.from_records(results)
df.to_csv('./mental_health.csv')
```
